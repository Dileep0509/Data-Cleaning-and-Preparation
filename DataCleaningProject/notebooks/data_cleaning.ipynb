## Total Price Validation

The `TotalPrice` column was validated by comparing it with the calculated value (`Quantity × UnitPrice`).

The initial comparison returned `False` because floating-point numbers are stored with small precision differences in computers.

After rounding both values to two decimal places, all records matched successfully.

**Result:** The `TotalPrice` values are correct.
